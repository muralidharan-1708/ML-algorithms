```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/Restaurant.xlsx')
df
```

```python
#Average sales on date 01.10.2023
a = df[df['DATE'] =='01.10.23']
a['AMOUNT'].mean()
```

```python
#Total amount spent by INVOICE_NO = 117
b = df[df['INVOICE_NO'] == 117]
print(b['AMOUNT'].sum())
```

```python
#Top 5 item sold
res = df.groupby('ITEM')['SUB_TOTAL'].sum().sort_values(ascending=False).head()
plt.bar(res.index, res.values)
plt.show()
```

```python
#Top 5 quantity of item sold
top_quantity = df.groupby('ITEM')['QUANTITY'].sum().sort_values(ascending=False).head()
plt.bar(top_quantity.index, top_quantity.values)
plt.show()
```
