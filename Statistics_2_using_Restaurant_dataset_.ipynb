```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('/content/Data.csv')
```

```python
df.columns
```

```python
df
```

```python
# highest average sale Date
amt = df.groupby('Date')['Final Total'].mean()
d = amt.sort_values(ascending=False).head(1)
d
```

```python
#total number of bills
bill = df.groupby('Invoice No.').count()
print(bill)
# df['Invoice No.'].nunique()
```

```python
# Most sold item of September month(quantity)?
month = df[(df['Date'] >= '2023-09-01') & (df['Date'] <= '2023-09-30')]
month
```

```python
month.groupby('Item Name')['Qty.'].sum().sort_values(ascending=False).head(1)
```

```python
# which 3 category item is sold less in the month of the july(Bar chart)?
m = df[(df['Date'] >= '2023-07-01') & (df['Date'] <= '2023-07-31')]
a = m.groupby('Category')['Qty.'].sum().sort_values(ascending=True).head(3)
a
plt.bar(a.index, a.values)
plt.show()
```

```python
# pie chart for sales by category
b = df.groupby('Category')['Final Total'].sum()
plt.pie(b, labels = b.index)
plt.show()
```

```python
# what is the Average of Value - total amount / Number of Bills?
t = df['Final Total'].sum()

b = df['Invoice No.'].nunique()

print(t/b)
```
