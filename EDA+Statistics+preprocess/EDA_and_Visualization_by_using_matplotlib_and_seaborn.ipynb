```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/diabetes.csv')
df.head()
# df[df.duplicated()]
```

```python
a = df['HbA1c_level']
b = df['age']
# scatter plot
plt.scatter(b,a)
plt.show()
```

```python
# Line plot
plt.plot(a)
```

```python
# Histogram
plt.hist(df['age'])
```

```python
plt.bar(df['age'],df['HbA1c_level'])
```

```python
plt.pie(df['diabetes'])
```

```python
# seaborn-lineplot
# all this work only for lineplot
sns.lineplot(a)
sns.lineplot(b)
sns.lineplot(x = a,y = b,data=df)
```

```python
# but in scatterplot use this format/ syntax
sns.scatterplot(x = a,y = b,data=df)
```

```python
# histogram without curve
sns.displot(df['age'])
```

```python
# histogram with curve
sns.distplot(df['age'])
```

```python
sns.barplot(x = a,y = b,data=df)
```

```python
# sns.boxplot(a)
# sns.boxplot(b)
sns.boxplot(x = a,y = b,data=df)
```

```python
sns.pairplot(df, hue = 'diabetes')
```

```python
c = df.corr()
sns.heatmap(c,annot=True)
```
